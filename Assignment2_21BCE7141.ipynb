# **Perform the Below Visualizations.**


*   **Univariate Analysis**
*   **Bi - Variate Analysis**
*   **Multivariate Analysis**

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("House Price India.csv")

# Univariate Analysis
plt.figure(figsize=(12, 4))
sns.histplot(data['living area'], kde=True)
plt.title('Univariate Analysis - Living Area')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.show()

# Bivariate Analysis
plt.figure(figsize=(12, 4))
sns.scatterplot(x='living area', y='Price', data=data)
plt.title('Bivariate Analysis - Living Area vs. Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()

# Multivariate Analysis
correlation_matrix = data.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Multivariate Analysis - Correlation Heatmap')
plt.show()
```
